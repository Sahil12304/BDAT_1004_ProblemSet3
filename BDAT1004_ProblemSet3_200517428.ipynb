```python
# the data starts from year 2061
import datetime
# data="C:/Users/19057/Documents/wind.txt"
# data2=pd.read_csv(data, sep='\s+' , parse_dates=[[0,1,2]])

def fix_century(t):
    year = t.year - 100 if t.year > 1989 else t.year
    return datetime.date(year, t.month, t.day)



data2['Yr_Mo_Dy'] = data2['Yr_Mo_Dy'].apply(fix_century)


data2.head()
    
    
```
